```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
monthly = pd.read_csv('monthly_kpis.csv')
pd.to_datetime(monthly.month)
monthly.set_index('month', inplace = True)

daily = pd.read_csv('daily_kpis.csv', index_col = 'day')
```

```python
# average number of patrons per check
monthly['avg_size'] = monthly.visitors/monthly.receipts
daily['avg_size_pre'] = daily.visitors_pre/daily.receipts_pre
daily['avg_size_post'] = daily.visitors_post/daily.receipts_post

# average number of items ordered per check
monthly['avg_order'] = monthly['items']/monthly.receipts
daily['avg_order_pre'] = daily.items_pre/daily.receipts_pre
daily['avg_order_post'] = daily.items_post/daily.receipts_post

# average amount of check
monthly['avg_check'] = monthly.revenue/monthly.receipts
daily['avg_check_pre'] = daily.revenue_pre/daily.receipts_pre
daily['avg_check_post'] = daily.revenue_post/daily.receipts_post

# average returns per patron
monthly['avg_per_cap'] = monthly.revenue/monthly.visitors
daily['avg_per_cap_pre'] = daily.revenue_pre/daily.visitors_pre
daily['avg_per_cap_post'] = daily.revenue_post/daily.visitors_post
```

```python
title_font = {'family' : 'serif',
              'size' : '30',
              'weight' : 'bold'}
```

```python
plt.figure(figsize = (20, 10))
plt.plot(monthly.visitors, '-bo')
plt.xlabel('Month')
plt.ylabel('Patrons')
plt.title(label = '# of patrons per month', fontdict = title_font)
plt.show() 

plt.figure(figsize = (20, 10))
plt.plot(monthly.avg_size, '-r*')
plt.xlabel('Month')
plt.ylabel('Average visitors per check')
plt.title(label = 'Avg. # of patrons per check', fontdict = title_font)
plt.show()
```

```python
plt.figure(figsize = (20, 10))
plt.plot(monthly.receipts, '-bo')
plt.xlabel('Month')
plt.ylabel('# of receipts')
plt.title(label = '# of receipts per month', fontdict = title_font)
plt.show() 

plt.figure(figsize = (20, 10))
plt.plot(monthly.avg_check, '-r*')
plt.xlabel('Month')
plt.ylabel('Average revenue per order in EUR')
plt.title(label = 'Avg. monthly per-check revenue (EUR)', fontdict = title_font)
plt.show()
```

```python
plt.figure(figsize = (20, 10))
plt.plot(monthly.revenue, '-bo')
plt.xlabel('Month')
plt.ylabel('Revenue in EUR')
plt.title(label = 'Revenue gained per month', fontdict = title_font)
plt.show() 

plt.figure(figsize = (20, 10))
plt.plot(monthly.avg_per_cap, '-r*')
plt.xlabel('Month')
plt.ylabel('Average revenue per patron')
plt.title(label = 'Avg. monthly per-capita revenue', fontdict = title_font)
plt.show()
```

```python
plt.figure(figsize = (20, 10))
plt.plot(monthly['items'], '-bo')
plt.xlabel('Month')
plt.ylabel('# of items sold')
plt.title(label = '# of items sold per month', fontdict = title_font)
plt.show() 

plt.figure(figsize = (20, 10))
plt.plot(monthly.avg_order, '-r*')
plt.xlabel('Month')
plt.ylabel('Average order size')
plt.title(label = 'Avg. order size per check', fontdict = title_font)
plt.show()
```
